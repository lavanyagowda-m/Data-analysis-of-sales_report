```python
import pandas as pd
lavan= pd.read_csv(r"C:\Users\Sharath G U\OneDrive\Desktop\lavan\data12.csv")
print(lavan.head())
print(lavan.info())   
print(lavan.describe()) 
product_sales = lavan.groupby("Product")["Sales"].sum()
print(product_sales)
region_sales =lavan.groupby("Region")["Sales"].sum()
print(region_sales)
import matplotlib.pyplot as plt
product_sales.plot(kind="bar", title="Sales by Product")
plt.show()
region_sales.plot(kind="pie", autopct="%1.1f%%", title="Sales by Region")
plt.ylabel("")  
plt.show()
```
